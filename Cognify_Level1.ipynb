Task 2: Gender Distribution

```python
import matplotlib.pyplot as plt
```

```python
gender_column = df['gender']
gender_column
```

```python
gender_counts = df.gender.value_counts()
gender_counts
```

The total count of Males and Females is 25 and 15 respectively. 

```python
plt.figure(figsize=(5, 6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'pink'],explode=[0,0.1],shadow=2, startangle=90)
plt.title('Distribution of Genders')
plt.ylabel('')
plt.show()
```

The dataset shows that the males occupy 62.5% of the total entries, while females occupy 37.5%.
